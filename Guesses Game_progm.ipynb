```python
from random import randint
number=randint(1,101)
#print(number)
guesses = []
while True:
    try:
        guess = int(input("Enter your guess (1-100): "))
    except:
        print("Please enter a valid integer.")
        continue
    if guess<1 or guess>100:
        print("OUT OF BOUNDS!")
        continue
    guesses.append(guess)
    if guess==number:
      print("Congratulations! You guessed it in ",len(guesses), "tries")
      break
    if len(guesses)==1:
        if abs(guess-number)<=10:
            print("Warm!")
        else:
            print("Cold!")
    else:
            if abs(guess-number)<abs(guesses[-2]-number):
                print("Warmer!")
            else:
                print("Colder!")

```
